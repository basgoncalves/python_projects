```python
import pandas as pd
from matplotlib import pyplot as plt

movies = pd.read_csv('titles.csv', encoding="ISO-8859-1")
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
movies_only_numeric = movies.select_dtypes(include=numerics)

print('size of dataset:')
print(len(movies) ,' x ', len(movies.columns))             # rows x columns
movies.head()
print('  ')
print(movies.dtypes)
print(movies[["release_year", "tmdb_score",'imdb_votes']].describe())
print('median value: ')
print('  ')
print('  ')
print('  ')
print('  ')
print(movies[["release_year", "tmdb_score",'imdb_votes']].median())

fig = plt.hist(movies.tmdb_score)
plt.show()

fig2 = plt.scatter(movies.tmdb_score,movies.imdb_votes)
plt.ylabel('imdb votes')
plt.xlabel('imdb score')
plt.show()

fig2 = plt.scatter(movies.release_year,movies.tmdb_score)
plt.ylabel('imdb_score')
plt.xlabel('release_year')
plt.show()


print('correlation matrix')
movies_only_numeric.corr()
```
